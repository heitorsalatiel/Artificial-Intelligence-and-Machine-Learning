## 3.2 Distribuição de gênero por departamento

Analise a distribuição de gênero por departamento. Mostre visualmente qual a proporção entre os gênero por departamento na empresa.

Você pode utilizar o [gráfico de barras](https://plotnine.readthedocs.io/en/stable/generated/plotnine.geoms.geom_col.html#two-variable-bar-plot) para fazer essa visualização. Sempre escreva as conclusões e insights obtidos a partir das análises.

```python
info_table = pd.concat([perfil_experiencia.area,perfil.genero], axis=1).reset_index()
info_table.drop(['perfil_id'],axis=1, inplace=True)

info_table['masculino'] = info_table.genero == 'Masculino'
info_table['feminino'] = info_table.genero == 'Feminino'
info_table['nao informaram'] = info_table.genero == 'Não informado'

info_table.drop(['genero'],axis=1,inplace=True)

grouped_info = info_table.groupby('area').agg({
    'masculino':'sum',
    'feminino':'sum',
    'nao informaram':'sum'
}).reset_index()

exp_gen_table = pd.melt(grouped_info,id_vars=['area'],var_name='genero',value_name='quantidade')

exp_gen_table.head()
```

```python
(ggplot(exp_gen_table,aes(x="area",y='quantidade',fill='genero')) +
geom_bar(stat='identity',position=position_dodge()) +
 theme(axis_text_x  = element_text(angle = 90, hjust = 1)) +
labs(title = "Distribuição de Gêneros por Departamento") +
ylab("Quantidade de Indivíduos") + 
xlab("Departamento"))
```

Pelo gráfico acima,é possível verificar que  os indivíduos que não informaram o gênero são do departamento de "CEO and Founder", "Commercial", "Data Science", "IT" e "Intern". Além disso, o número de homens tende a ser maior do que o número de mulheres na maioria dos departamentos. Departamentos como "Education" e "Writer" só possuem mulheres, enquantos que o "Engineering" e o "Operational" só possuem homens. Há alguns departamentos que possuem o mesmo número de homens e de mulheres, como é o caso para os departamentos: "Financial", "Marketing"e "Design"

## 3.4 Diferenças na formação acadêmica por gênero

Analise a diferenças na formação acadêmica por gênero.

```python
formacao_genero = pd.concat([perfil_formacao.area,perfil.genero], axis=1).reset_index()
formacao_genero.drop(['perfil_id'],axis=1, inplace=True)

formacao_genero['masculino'] = formacao_genero.genero == 'Masculino'
formacao_genero['feminino'] = formacao_genero.genero == 'Feminino'
formacao_genero['nao informaram'] = formacao_genero.genero == 'Não informado'

formacao_genero.drop(['genero'],axis=1,inplace=True)

grouped_info = formacao_genero.groupby('area').agg({
    'masculino':'sum',
    'feminino':'sum',
    'nao informaram':'sum'
}).reset_index()

form_gen_table = pd.melt(grouped_info,id_vars=['area'],var_name='genero',value_name='quantidade')

form_gen_table.head()
```

```python
(ggplot(form_gen_table,aes(x="area",y='quantidade',fill='genero')) +
geom_bar(stat='identity',position=position_dodge()) +
 theme(axis_text_x  = element_text(angle = 90, hjust = 1)) +
labs(title = "Distribuição de Gêneros por Formação") +
ylab("Quantidade de Indivíduos") + 
xlab("Área"))
```

Pelo gráfico acima é possível perceber que as áreas de Administração, Marketing e RH, tem uma predominância do público feminino, sendo que no RH não há a presença de nenhum indivíduo masculino.  De forma análoga nas categorias Engenharia, TI e Outros há uma forte predominância do público masculino. Na categoria Direito e Economia só há a presença de homens. Além disso, os indivíduos que não informaram o sexo pertencem aos setores de Administração, Economia, Engenharia, Marketing, Outros e TI. É possível perceber então, que há uma grande tendência das áreas de tecnologia serem populadas com membros do sexo masculino, enquanto que áreas de gestão, marketing e administração de pessoal, serem populadas com membros do sexo feminino.

## 3.5 Multidisciplinaridade nos departamentos

Mostre o perfil acadêmico nos diferentes setores/departamentos da empresa.

```python
experiencia_formacao = perfil_experiencia.merge(perfil_formacao, how='inner',on='perfil_id')
experiencia_formacao.head()
```

```python
exp_form = pd.concat([perfil_experiencia.area,perfil_formacao.area], axis=1).reset_index()
exp_form.drop(['perfil_id'],axis=1, inplace=True)
exp_form.columns=['area_exp','area_form']
areas = exp_form.area_exp.unique()
rows = []
for area in areas:
    cond = exp_form.area_exp == area
    selected = exp_form[cond]
    values = selected.area_form.value_counts()
    for key in values.keys():
        rows.append([key,area,values[key]])
exp_form_table = pd.DataFrame(rows,columns=['area_form','area_exp','quantidade'])
exp_form_table.head()
```

```python
(ggplot(exp_form_table,aes(x="area_exp",y='quantidade',fill='area_form')) +
geom_bar(stat='identity',position=position_dodge()) +
 theme(axis_text_x  = element_text(angle = 90, hjust = 1)) +
labs(title = "Distribuição de Formação por Departamento") +
ylab("Quantidade de Cursos") + 
xlab("Departamento"))
```

Pelo Gráfico acima é possível perceber que em diversos departamentos há uma mescla de formação. O departamento de Engenharia é formado por membros de TI, provavelmente por engenheiros de softwares. As formações em TI e Engenharia estão distribuídas em várias áreas, enquanto que nos departamentos de Marketing, Comercial e CEO's há uma forte concentração de indivíduos formados em áreas administrativas.

## 3.6 Escolaridade na empresa e nos departamentos


```python
perfil_formacao.head()
```

```python
area_counts = perfil_formacao.area.value_counts()
area_counts
```

```python
form_area_group = perfil_formacao.groupby('area').agg({
    'duracao':'sum'
}).reset_index()
form_area_group['mean_time'] = form_area_group.apply(lambda x: x.duracao / area_counts[x.area], axis=1)
form_area_group.head()
```

```python
(ggplot(form_area_group,aes(x="area",y='mean_time',fill='area')) +
geom_bar(stat='identity',position=position_dodge()) +
theme(axis_text_x  = element_text(angle = 90, hjust = 1)) +
labs(title = "Área de Formação x Tempo Médio de Formação") +
ylab("Tempo Médio") + 
xlab("Área de Formação"))
```

No gráfico acima é possível ver que o tempo médio na duração da formação é em Direito, porém só há uma pessoa com essa formação. Em segundo lugar vem os Engenheiros, que correspondem a 19 indivíduos na empresa. O menor tempo médio é em Data Science, porém só há um indivíduo com essa formação.

```python
perfil_formacao.grau.value_counts()
```

Na empresa há 78 indivíduos com graduação, 22 que possuem pós graduação, 5 que possuem mestrado 2 técnicos e 1 do ensino médio

```python
perfil_formacao['mestrado'] = perfil_formacao.grau == 5
perfil_formacao['posgraduacao'] = perfil_formacao.grau == 4
perfil_formacao['graduacao'] = perfil_formacao.grau == 3
perfil_formacao['tecnico'] = perfil_formacao.grau == 2
perfil_formacao['medio'] = perfil_formacao.grau == 1

perfil_formacao['mestrado'] = perfil_formacao['mestrado'].astype(int)
perfil_formacao['posgraduacao'] = perfil_formacao['posgraduacao'].astype(int)
perfil_formacao['graduacao'] = perfil_formacao['graduacao'].astype(int)
perfil_formacao['tecnico'] = perfil_formacao['tecnico'].astype(int)
perfil_formacao['medio'] = perfil_formacao['medio'].astype(int)

perfil_formacao.head()
```

```python
group_area = perfil_formacao.groupby('area').agg({
    'mestrado':'sum',
    'posgraduacao':'sum',
    'graduacao':'sum',
    'tecnico':'sum',
    'medio':'sum'
}).reset_index()
group_area.head()
```

```python
melt_area_table = pd.melt(group_area,id_vars=['area'],var_name='nivel',value_name='nivel_qtd')
melt_area_table.head()
```

```python
(ggplot(melt_area_table,aes(x="area",y='nivel_qtd',fill='nivel')) +
geom_bar(stat='identity',position=position_dodge()) +
 theme(axis_text_x  = element_text(angle = 90, hjust = 1)) +
labs(title = "Relação Formação e Nível de Formação") +
ylab("Quantidade Nível") + 
xlab("Área"))
```

Pelo gráfico acima é possível perceber que maioria dos indivíduos da graduação são da Engenharia. Já a maioria dos indivíduos que possuem pós graduação são formados em áreas administrativas. Já os indivíduos que possuem Mestrado são de "TI", "Outros" e "Marketing"

## 3.10 Análises Adssionais

Faça análises que julgar necessário para refutar ou corraborar a hipotése de Aquiles.

```python
perfil_habilidades.head()
```

```python
gender_formacao = perfil_habilidades.merge(perfil,how='inner',on='perfil_id').merge(perfil_formacao,how='inner',on='perfil_id')\
                  .loc[:,['area','genero','qtd_habilidades']]
gender_formacao['feminino'] = gender_formacao.genero=='Feminino'
gender_formacao['masculino'] = gender_formacao.genero=='Masculino'
gender_formacao['qtd_masc'] = gender_formacao.apply(lambda x: x.qtd_habilidades if x.genero == 'Masculino' else 0, axis=1)
gender_formacao['qtd_fem'] = gender_formacao.apply(lambda x: x.qtd_habilidades if x.genero == 'Feminino' else 0, axis=1)
gender_formacao = gender_formacao.groupby('area').agg({
    'feminino':'sum',
    'masculino':'sum',
    'qtd_masc':'sum',
    'qtd_fem':'sum'
}).reset_index()
gender_formacao['qtd_masc'] = gender_formacao['qtd_masc'] / gender_formacao['masculino']
gender_formacao['qtd_fem'] = gender_formacao['qtd_fem'] / gender_formacao['feminino']
gender_formacao.fillna(0, inplace=True)
gender_formacao.head()
```

```python
gender_formacao_melt = pd.melt(gender_formacao,id_vars=['area'],value_vars=['qtd_masc','qtd_fem'],)
gender_formacao_melt.head()
```

```python
(ggplot(gender_formacao_melt,aes(x="area",y='value', fill='variable')) +
geom_col(stat='identity',position=position_dodge()) +
theme(axis_text_x  = element_text(angle = 90, hjust = 1)) +
labs(title = "Relação Quantidade Média de Habilidades por Formação e Gênero") +
ylab("Quantidade Média de Habilidades") + 
xlab("Área de Formação"))
```

O gráfico acima nos entrega a informação a respeito da quantidade habilidades média por gênero e por área de formação. É possível ver que em *Administração* e *Engenharia*, os homens possuem em média menor quantidade de habilidades do que as mulheres, enquanto que em *Design*, *Finanças* e *TI* as habilidades dos homens é em média maior em quantidade do que a das mulheres.
